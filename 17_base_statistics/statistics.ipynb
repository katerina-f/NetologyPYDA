##### Задание 1. Базовое изучение
Изучить представленный набор данных на основе описания его столбцов и выбрать 8 столбцов для дальнейшего изучения (среди них должны быть как числовые, так и категориальные). Провести расчет базовых метрик для них, кратко описать результаты.


```python
# посмотрим на данные в столбцах 
horse_data = pd.read_csv("horse_data.csv")
horse_data.columns = ["surgery", "age", "hospital_number", "rectal_temperature", "pulse", "respiratory_rate", 
                      "temperature_of_extremities", "peripheral_pulse", "mucous membranes", 
                      "capillary_refill_time", "pain", "peristalsis", "abdominal_distension", 
                      "nasogastric_tube","nasogastric_reflux", "nasogastric_reflux_PH", "rectal_examination", 
                      "abdomen", "packed_cell_volume", "total_protein", "abdominocentesis_appearance", 
                     "abdomcentesis_total_protein", "outcome", "surgical_lesion?", 
                      "first_number_is_site_of_lesion", "second_number_is_type", 
                      "third_number_is_subtype", "cp_data"]
horse_data
```

```python
# выберем восемь признаков для анализа 
eight_horse_data = horse_data[['age', 'rectal_temperature', 'pulse', 'temperature_of_extremities', 
                               'capillary_refill_time', 'outcome', 'abdominal_distension',
                               'peristalsis']]

# встречается некоторое количество знаков вопроса - заменим на NaN и приведем к числу все значения
eight_horse_data = eight_horse_data.replace('?', np.nan).apply(lambda column: pd.to_numeric(column), axis=0)
eight_horse_data.head()
```

```python
eight_horse_data.describe()
```

Выводы по первому взгляду на базовые метрики:
    - мы видим, что пульс больной лошади в среднем превышает пульс здоровой на 30 ударов в минуту 
    
    - субъективная оценка температуры конечностей по большей части - конечности ближе к холодным (прохладные), что говорит о шоковом состоянии 
    
    - температура конечностей коррелирует с проходимостью сосудов - чем проходимость ниже, тем холоднее конечности лошади
    
    - вздутие живота на 75 перцентиле умеренное 
    
    - предыдущий параметр коррелирует со вздутием - низкая проходимость кишечника зарегистрирована у большинства лошадей, так же как и повышенное вздутие живота, показатели связаны между собой

##### Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
# расчет выбросов по пульсу 
pulse_iqr = eight_horse_data.pulse.quantile(0.75) - eight_horse_data.pulse.quantile(0.25)
pulse_lower_bound = eight_horse_data.pulse.quantile(0.25) - (1.5 * pulse_iqr) 
pulse_upper_bound = eight_horse_data.pulse.quantile(0.75) + (1.5 * pulse_iqr)
print(f"Нижняя граница пульса: {pulse_lower_bound}")
print(f"Верхняя граница пульса: {pulse_upper_bound}")

print(f"Выбросы по пульсу: \nНижние:\n{eight_horse_data.loc[eight_horse_data.pulse < pulse_lower_bound].pulse}\n\nВерхние:\n{eight_horse_data.loc[eight_horse_data.pulse > pulse_upper_bound].pulse}")
```

Видим, что нижней границей является отрицательный пульс - этого не может быть по определению, но число мало, поэтому из наших вычислений видим, что нижних выбросов нет и отрицательных значений пульса тоже

```python
print(f"Нижняя граница пульса: {eight_horse_data.pulse.quantile(0.25)}")
```

Посмотрев на верхние выбросы, можем сказать, что из общих данных сильно выбивается лишь одно значение, взглянем на 75 перцентиль

```python
print(f"Верхняя граница пульса: {eight_horse_data.pulse.quantile(0.75)}")
```

Это значение очень отличается от расчетов плюс минус 1.5 - почти в два раза, можно посмотреть, сколько значений выбивается за пределы 75 перцентиля


```python
eight_horse_data.loc[eight_horse_data.pulse > eight_horse_data.pulse.quantile(0.75)].pulse
```

Значений, превышающих 75 перцентиль очень много - 67, можем сделать вывод, что они попали в выборку не случайно и важны для вычислений, поэтому уберем только те значения, которые превышают нашу верхнюю границу

```python
# расчет выбросов по ректальной температуре
temp_iqr = eight_horse_data.rectal_temperature.quantile(0.75) - eight_horse_data.rectal_temperature.quantile(0.25)
temp_lower_bound = eight_horse_data.rectal_temperature.quantile(0.25) - (1.5 * temp_iqr) 
temp_upper_bound = eight_horse_data.rectal_temperature.quantile(0.75) + (1.5 * temp_iqr)
print(f"Нижняя граница температуры: {temp_lower_bound}")
print(f"Верхняя граница температуры: {temp_upper_bound}")

print(f"Выбросы по температуре: \nНижние:\n{eight_horse_data.loc[eight_horse_data.rectal_temperature < temp_lower_bound].rectal_temperature}\n\nВерхние:\n{eight_horse_data.loc[eight_horse_data.rectal_temperature > temp_upper_bound].rectal_temperature}")
```

По ректальной температуре мы видим, что разница между нижней границей и минимальным значением, которое попало в выбросы очень мала - то есть можем сделать вывод, что это значение как таковое выбросом не является и его можно учитывать в общем анализе 

```python
# убираем выбросы
eight_horse_data = eight_horse_data[eight_horse_data.pulse.between(pulse_lower_bound, pulse_upper_bound, inclusive=True)]
eight_horse_data.describe()
```

##### Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
for col in eight_horse_data.columns:
    pct_missing = eight_horse_data[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

Мы видим пропуски в столбцах temperature_of_extremities, abdominal_distension, rectal_temperature, peristalsis - их не большое количество. В остальных столбцах пропусков мало или нет, поэтому можно считать, что мы можем их игнорировать 

```python
# ректальная температура - числовое значение, заменим его средним значением
eight_horse_data.rectal_temperature.fillna(eight_horse_data.rectal_temperature.mean(), inplace=True)
eight_horse_data.rectal_temperature.describe()
```

Видим, что на общих данных замена пустых значений ректальной температуры практически не сказалась 

```python
# температура конечностей - категориальная величина, ее обычно заменяют модой
eight_horse_data.temperature_of_extremities.fillna(eight_horse_data.temperature_of_extremities.mode(), inplace=True)
eight_horse_data.temperature_of_extremities.describe()
```

Здесь мы можем увидеть, что при заполнении пропусков среднее значение и отклонение также незначительно скорректировались - среднее увеличилось, а отклонение напротив уменьшилось

```python
# вздутие живота также величина категориальная - заполняем модой 
eight_horse_data.abdominal_distension.fillna(eight_horse_data.abdominal_distension.mode(), inplace=True)
eight_horse_data.abdominal_distension.describe()
```

Со значениями по вздутию живота произошло то же, что и с температурой конечностей - мы уменьшили отклонение, но несколько сильнее

```python
# проходимость кишечника так же категориальная - заполняем модой
eight_horse_data.peristalsis.fillna(eight_horse_data.peristalsis.mode(), inplace=True)
eight_horse_data.peristalsis.describe()
```

В случае с peristalsis данные почти не изменились, однако 75 перцентиль сместился на значение "умеренной проходимости"
