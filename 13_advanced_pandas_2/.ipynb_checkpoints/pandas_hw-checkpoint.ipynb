#### Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
df_rat = pd.read_csv("ml-latest-small/ratings.csv")
```

```python
def count_rating_and_timestamp(row):
    row["ratingCount"] = row.rating.count()
    row["lifeTime"] = row.timestamp.max() - row.timestamp.min()
    return row

data = df_rat.groupby("userId").apply(count_rating_and_timestamp)[["ratingCount", "lifeTime"]].drop_duplicates()
round(data.loc[data.ratingCount > 100].lifeTime.mean(), 2)
```

#### Задание 2

Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). Необходимо сформировать две таблицы:
- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# таблица с тремя типами перевозок 
import numpy as np

svod = (
    client_base.merge(rzd, on="client_id", how="left").
    merge(auto, on="client_id", how="left").
    merge(air, on="client_id", how="left").replace(np.nan, 0)
)

# с адресом
svod
```

```python
# без адреса 

svod.drop("address", axis=1)
```
