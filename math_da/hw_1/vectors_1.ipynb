#### Уровень 0:
Даны вектора x=(1,1), y=(2, 0) и z = (0,2)

Надо:

- Изобразить вектора на экране
- Изобразить точку x + y + z
- Найти угол между векторами x и y

```python
x = np.array([1, 1])
y = np.array([2, 0])
z = np.array([0, 2])

n = x + y + z

plt.figure()
plt.plot(x[0], x[1], 'o')
plt.plot(y[0], y[1], 'o')
plt.plot(z[0], z[1], 'o')
plt.plot(n[0], n[1], 'o')
plt.grid()
```

```python
len_x = np.linalg.norm(x)
len_y = np.linalg.norm(y)

cos = np.dot(x, y)/np.dot(len_x, len_y)

angle = np.degrees(np.arccos(cos))
angle
```

#### Уровень 1:
Даны дома из Бостона

- Выбрать любые n домов, образующих базис в пространстве признаков дома (все колонки, кроме medv)

- Написать код разложения дома по указанному базису

```python
boston_data = pd.read_csv('boston_data.csv')
boston_data.drop("medv", axis=1, inplace=True)
houses = boston_data.dropna()
houses
```

```python
houses = np.array(houses)
basis = []
for i in range(np.linalg.matrix_rank(houses, tol=None)):
    basis.append(list(houses[i]))

basis = np.array(basis)
np.linalg.solve(basis.transpose(), houses[36])
```
