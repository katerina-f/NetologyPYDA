Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

```python
def draw_func(func, mu, sigma):
    if func == sts.t:
        func_rv = func(sigma)
    elif func == sts.expon:
        func_rv = func(mu, scale=sigma)
    else:
        kwargs = {"loc": mu, "scale": sigma}
        func_rv = func(**kwargs)
    vector = np.linspace(-5,5,100)
    func_rv.rvs(10)
    pdf = func_rv.pdf(vector)
    plt.plot(vector, pdf)
```

```python
#нормальное
for i in range(5):
    mu = i
    sigma = 1/1+i
    draw_func(sts.norm, mu, sigma)
    
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
#экспоненциальное
for i in range(5):
    mu = i
    sigma = 1/1+i
    draw_func(sts.expon, mu, sigma)
    
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```
