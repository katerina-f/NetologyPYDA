```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

#### Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции

`f(x) = x^5 + 4sin(2x) + cos(3x+3)` в точке `x = 1`


```python
def func(x):
    y = x*5 + 4*np.sin(x*2) + np.cos(3*x+3) 
    return y

diff_1 = derivative(func, 1, dx=1e-6)
diff_2 = derivative(func, 1, dx=1e-6, n=2)

diff_1, diff_2
```
